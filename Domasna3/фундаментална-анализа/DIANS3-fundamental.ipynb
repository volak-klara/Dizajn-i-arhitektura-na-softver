```python
from textblob import TextBlob

df = pd.read_csv('companyNews.csv', encoding='utf-8')

financial_sentiment = {
   # Positive terms
   'profit': 0.8,
   'growth': 0.7,
   'increase': 0.6,
   'dividend': 0.7,
   'expansion': 0.5,
   'success': 0.4,
   'positive': 0.4,
   'investment': 0.5,
   'development': 0.5,
   'improvement': 0.4,
   
   # Neutral terms
   'report': 0.0,
   'statement': 0.0,
   'meeting': 0.0,
   'announcement': 0.1,
   'information': 0.0,
   
   # Negative terms
   'loss': -0.8,
   'decrease': -0.6,
   'decline': -0.6,
   'debt': -0.5,
   'risk': -0.5,
   'unaudited': -0.2,
   'price sensitive': -0.3,
   'delay': -0.4,
   'penalty': -0.7,
   'issue': -0.5
}

def analyze_sentiment(text):
   if not isinstance(text, str):
       return 0
   
   text = text.lower()
   blob = TextBlob(text)
   base_sentiment = blob.sentiment.polarity
   
   financial_score = 0
   matches = 0
   for term, value in financial_sentiment.items():
       if term in text:
           financial_score += value
           matches += 1
   
   financial_score = financial_score / max(matches, 1) if matches > 0 else 0
   return (base_sentiment + financial_score) / 2

def get_recommendation(score):
   if score > 0.3:
       return "BUY"
   elif score < -0.3:
       return "SELL"
   return "HOLD"

df['sentiment_score'] = df['Title'].apply(analyze_sentiment)


company_analysis = df.groupby('Company').agg({
   'sentiment_score': ['mean', 'count'],
   'Date': ['min', 'max']
}).reset_index()

company_analysis.columns = ['Company', 'Avg_Sentiment', 'News_Count', 'First_News', 'Last_News']


company_analysis['Code'] = company_analysis['Company'].apply(
    lambda x: next((code for code, name in company_codes.items() if x.startswith(name)), 'N/A')
)
company_analysis['Recommendation'] = company_analysis['Avg_Sentiment'].apply(get_recommendation)


company_analysis = company_analysis[['Code', 'Company', 'Avg_Sentiment', 'News_Count', 'First_News', 'Last_News', 'Recommendation']]
company_analysis = company_analysis.sort_values('Avg_Sentiment', ascending=False)

# Print results
print("\nFundamental Analysis Based on News Sentiment:")
print("=" * 80)
for _, row in company_analysis.iterrows():
   print(f"\nCode: {row['Code']}")
   print(f"Company: {row['Company']}")
   print(f"Average Sentiment: {row['Avg_Sentiment']:.2f}")
   print(f"News Count: {row['News_Count']}")
   print(f"Analysis Period: {row['First_News']} to {row['Last_News']}")
   print(f"Recommendation: {row['Recommendation']}")

# Save results
company_analysis.to_csv('fundamental_analysis_results_en.csv', index=False, encoding='utf-8')
```
